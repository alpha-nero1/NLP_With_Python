```python
# Calc a new vector by adding and subtracting vectors
from scipy import spatial
```

```python
cosine_sim = lambda vec1, vec2: 1 - spatial.distance.cosine(vec1, vec2)
```

```python
king = nlp.vocab['king'].vector
man = nlp.vocab['man'].vector
wo_man = nlp.vocab['woman'].vector
```

```python
# king - man + woman
new_vector = king - man + wo_man
```

```python
new_vector
```

```python
computed_similarities = []

# Find similar words to the vector!!
for word in nlp.vocab:
    if (word.has_vector and word.is_lower and word.is_alpha):
        sim = cosine_sim(new_vector, word.vector)
        computed_similarities.append((word, sim))
```

```python
# Sort by similarity descending
computed_similarities = sorted(computed_similarities, key=lambda item: -1 * item[1])
```
